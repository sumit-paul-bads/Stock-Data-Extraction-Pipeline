```python
import requests
import pandas as pd
import time
```

## **Defining Fetch Function**

```python
def fetch_stock_data(symbol):
    url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey={API_KEY}&outputsize=compact"

    response = requests.get(url)
    data = response.json()

    print("FULL RESPONSE:", data)

    if "Time Series (Daily)" not in data:
        print(f"Error for {symbol}")
        return None

    df = pd.DataFrame.from_dict(data["Time Series (Daily)"], orient='index')

    df.columns = ["open", "high", "low", "close", "volume"]
    df.index = pd.to_datetime(df.index)
    df = df.sort_index()

    return df
```

### **Fetching NSE**

```python
df_nse = fetch_stock_data("HDB")
df_nse.head()
```

```python
df_bse = fetch_stock_data("HDFCBANK.BSE")
df_bse.head()
```

```python
print (df_nse)
```

Indian stock market is filled with holidays and stuff.

So we create a complete calender range from start date to end date then reindex the stock data into new timeline.

```python
start_date = df_nse.index.min()
end_date = df_nse.index.max()

full_date_range = pd.date_range(start=start_date, end=end_date, freq='D')

df_nse_reindexed = df_nse.reindex(full_date_range)
df_bse_reindexed = df_bse.reindex(full_date_range)

df_nse_cleaned = df_nse_reindexed.ffill()
df_bse_cleaned = df_bse_reindexed.ffill()

print("df_nse_cleaned:")
print(df_nse_cleaned)

print("\ndf_bse_cleaned")
print(df_bse_cleaned)
```

## **Saving to CSV**

```python
df_nse_cleaned.to_csv("nse_data.csv")
df_bse_cleaned.to_csv("bse_data.csv")
```
